### cleaning up spelling mistakes


```python
import nltk
from nltk.corpus import wordnet
from spellchecker import SpellChecker
import requests

# Download WordNet data
nltk.download('wordnet', quiet=True)

# Initialize SpellChecker
spell = SpellChecker()

# Load additional word lists
def load_words_from_file(filename):
    try:
        with open(filename, 'r') as file:
            return file.read().splitlines()
    except FileNotFoundError:
        print(f"Warning: {filename} not found. Skipping.")
        return []

# Load words from multiple files
for filename in ['english_words.txt', 'technical_terms.txt', 'proper_nouns.txt']:
    spell.word_frequency.load_words(load_words_from_file(filename))

# Function to check if a word is in WordNet
def is_english_word(word):
    return len(wordnet.synsets(word)) > 0

# Function to check word in online dictionary (Free Dictionary API)
def check_word_in_dictionary(word):
    url = f"https://api.dictionaryapi.dev/api/v2/entries/en/{word}"
    try:
        response = requests.get(url)
        return response.status_code == 200
    except:
        return False

def correct_spelling(text, use_api=False):
    words = text.split()
    corrected_words = []
    uncorrectable_words = {}
    
    for word in words:
        lower_word = word.lower()
        if lower_word in ['ths', 'th']:
            corrected_words.append('this')
        elif is_english_word(word) or lower_word in spell:
            corrected_words.append(word)
        elif use_api and check_word_in_dictionary(word):
            corrected_words.append(word)
        else:
            # Check if it's a compound word
            if any(char.isdigit() for char in word):
                parts = [''.join(filter(str.isalpha, word)), ''.join(filter(str.isdigit, word))]
                if all(is_english_word(part) or part.lower() in spell for part in parts):
                    corrected_words.append(word)
                    continue
            
            corrected_word = spell.correction(word)
            if corrected_word == word or not corrected_word:
                corrected_words.append(word)
                uncorrectable_words[word] = corrected_word
            else:
                corrected_words.append(corrected_word)
    
    return ' '.join(corrected_words), uncorrectable_words

# Example usage
if __name__ == "__main__":
    example_text = "Ths is an exmple of custmword1 and sme technical jargn."
    corrected_text, uncorrectable = correct_spelling(example_text, use_api=False)
    
    print("Original Text:", example_text)
    print("Corrected Text:", corrected_text)
    print("Uncorrectable Words:", uncorrectable)

    # Using the Free Dictionary API
    corrected_text_api, uncorrectable_api = correct_spelling(example_text, use_api=True)
    print("\nCorrected Text (with API):", corrected_text_api)
    print("Uncorrectable Words (with API):", uncorrectable_api)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans
from kneed import KneeLocator

grouped = filler_data_noms.groupby('Reason / Fault')
results = {}

for name, group in grouped:
    if len(group) < 5:
        print(f"Skipping group {name} due to insufficient group size.")
        continue

    vectorizer = TfidfVectorizer(sublinear_tf=True, min_df=1, max_df=1.0)
    
    try:
        X = vectorizer.fit_transform(group['lemmatized']).toarray()
    except ValueError:
        print(f"Skipping group {name} due to insufficient data after vectorization.")
        continue

    if X.shape[1] < 2:
        print(f"Skipping group {name} due to insufficient features for clustering.")
        continue

    max_clusters = min(len(group), 10)
    if max_clusters < 2:
        print(f"Not enough data to form multiple clusters in group {name}.")
        continue

    inertias = []
    cluster_options = range(1, max_clusters + 1)
    for num_clusters in cluster_options:
        kmeans = KMeans(n_clusters=num_clusters, random_state=42)
        kmeans.fit(X)
        inertias.append(kmeans.inertia_)

    if len(cluster_options) > 1:
        knee = KneeLocator(cluster_options, inertias, curve='convex', direction='decreasing')
        optimal_clusters = knee.knee if knee.knee else 1
    else:
        optimal_clusters = 1

    plt.figure()
    plt.plot(cluster_options, inertias, marker='o')
    plt.title(f'Elbow Method For Optimal k in Group: {name}')
    plt.xlabel('Number of clusters')
    plt.ylabel('WCSS (Inertia)')
    plt.grid(True)
    plt.plot(optimal_clusters, inertias[cluster_options.index(optimal_clusters)], 'ro')  # Red marker at the elbow
    plt.show()

    kmeans = KMeans(n_clusters=optimal_clusters, random_state=42)
    kmeans.fit(X)

    group['Cluster'] = kmeans.labels_
    results[name] = {
        'clusters': kmeans.labels_,
        'centers': kmeans.cluster_centers_,
        'terms': vectorizer.get_feature_names_out()
    }

    filler_data_noms.loc[group.index, 'Cluster'] = kmeans.labels_
```

### trying out the cluster prediction with input texts

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.ensemble import RandomForestClassifier
import joblib




# Function to predict cluster for new input text
def predict_cluster(input_text):
    # Load the model and vectorizer
    model = joblib.load('random_forest_model.pkl')
    vectorizer = joblib.load('tfidf_vectorizer.pkl')
    
    # Preprocess the input text
    input_text_processed = input_text.lower()
    
    # Transform the input text using the TF-IDF vectorizer
    input_features = vectorizer.transform([input_text_processed])
    
    # Predict the cluster using the trained model
    predicted_cluster = model.predict(input_features)
    
    return predicted_cluster[0]

# Example usage
input_text = "there seems to be an issue with the lubrication roller failure that has caused jammed cans"
predicted_cluster = predict_cluster(input_text)
print(f"The predicted cluster is: {predicted_cluster}")
```

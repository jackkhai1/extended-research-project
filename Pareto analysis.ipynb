### pareto chart for the area/ cause column

```python
import pandas as pd
import matplotlib.pyplot as plt

# Example DataFrame setup
# Assuming you have a DataFrame 'filler_data' with a column 'Area / Cause'
# Example: filler_data = pd.DataFrame({'Area / Cause': ['Cause1', 'Cause2', 'Cause1', ...]})

# Count the occurrences of each 'Area / Cause'
cause_counts = filler_data['Area / Cause'].value_counts()

# Sort the causes in descending order
sorted_counts = cause_counts.sort_values(ascending=False)

# Calculate the cumulative percentage
cumulative_percentage = sorted_counts.cumsum() / sorted_counts.sum() * 100

# Create the Pareto chart
fig, ax1 = plt.subplots(figsize=(10, 6))

# Plot the bar chart
sorted_counts.plot(kind='bar', color='skyblue', ax=ax1)
ax1.set_xlabel('Area / Cause')
ax1.set_ylabel('Frequency')
ax1.set_title('Pareto Chart of Area / Cause')

# Plot the cumulative percentage line on a secondary y-axis
ax2 = ax1.twinx()
cumulative_percentage.plot(marker='o', color='orange', ax=ax2)
ax2.set_ylabel('Cumulative Percentage')

# Add a horizontal line for the 80% threshold
ax2.axhline(80, color='red', linestyle='--')
ax2.annotate('80% Threshold', xy=(0.5, 80), xytext=(5, 80),
             textcoords='offset points', color='red')

plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

## pareto chart for the reason/fault column

```python
import pandas as pd
import matplotlib.pyplot as plt

# Example DataFrame setup
# Assuming you have a DataFrame 'filler_data' with a column 'Reason / Fault'
# Example: filler_data = pd.DataFrame({'Reason / Fault': ['Fault1', 'Fault2', 'Fault1', ...]})

# Count the occurrences of each 'Reason / Fault'
fault_counts = filler_data['Reason / Fault'].value_counts()

# Sort the faults in descending order
sorted_faults = fault_counts.sort_values(ascending=False)

# Calculate the cumulative percentage
cumulative_percentage_faults = sorted_faults.cumsum() / sorted_faults.sum() * 100

# Create the Pareto chart
fig, ax1 = plt.subplots(figsize=(10, 6))

# Plot the bar chart
sorted_faults.plot(kind='bar', color='lightgreen', ax=ax1)
ax1.set_xlabel('Reason / Fault')
ax1.set_ylabel('Frequency')
ax1.set_title('Pareto Chart of Reason / Fault')

# Plot the cumulative percentage line on a secondary y-axis
ax2 = ax1.twinx()
cumulative_percentage_faults.plot(marker='o', color='orange', ax=ax2)
ax2.set_ylabel('Cumulative Percentage')

# Add a horizontal line for the 80% threshold
ax2.axhline(80, color='red', linestyle='--')
ax2.annotate('80% Threshold', xy=(0.5, 80), xytext=(5, 80),
             textcoords='offset points', color='red')

plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Group the data by 'Area / Cause' and 'Reason / Fault', and count occurrences
grouped_data = filler_data.groupby(['Area / Cause', 'Reason / Fault']).size().unstack(fill_value=0)

# Generate a color palette
color_palette = sns.color_palette("husl", n_colors=len(grouped_data.columns))

# Create a Pareto chart for each 'Area / Cause'
for area in grouped_data.index:
    plt.figure(figsize=(14, 8))
    
    # Extract the data for the current area
    data_for_area = grouped_data.loc[area]
    
    # Remove columns with zero counts and sort in descending order
    data_for_area = data_for_area[data_for_area > 0].sort_values(ascending=False)
    
    # Calculate cumulative percentage
    cumulative_percentage = data_for_area.cumsum() / data_for_area.sum() * 100
    
    # Create Pareto Chart
    ax1 = plt.gca()
    ax2 = ax1.twinx()
    
    # Plot bars
    bars = ax1.bar(data_for_area.index, data_for_area.values, color=color_palette)
    
    # Plot cumulative percentage line
    ax2.plot(data_for_area.index, cumulative_percentage, color='red', marker='D', ms=7)
    
    # Customize the chart
    ax1.set_xlabel('Reason / Fault')
    ax1.set_ylabel('Frequency')
    ax2.set_ylabel('Cumulative Percentage')
    plt.title(f'Pareto Chart of Reason / Fault for {area}')
    
    # Rotate x-axis labels
    plt.xticks(rotation=45, ha='right')
    
    # Set y-axis limits for percentage
    ax2.set_ylim([0, 110])
    
    # Add percentage labels on the right y-axis
    ax2.set_yticks([0, 20, 40, 60, 80, 100])
    ax2.set_yticklabels(['0%', '20%', '40%', '60%', '80%', '100%'])
    
    # Add value labels on top of bars
    for bar in bars:
        height = bar.get_height()
        ax1.text(bar.get_x() + bar.get_width()/2., height,
                 f'{height:.0f}',
                 ha='center', va='bottom')
    
    # Add legend
    lines, labels = ax1.get_legend_handles_labels()
    lines2, labels2 = ax2.get_legend_handles_labels()
    ax2.legend(lines + lines2, labels + labels2, loc='upper left', bbox_to_anchor=(1, 1))
    
    plt.tight_layout()
    plt.show()
```
